```python
import numpy as np
import matplotlib.pyplot as plt

dx = 0.01
x = np.arange(1,10,1/100)
plt.plot(x, np.gradient(np.sin(x), dx))
plt.plot(x, np.cos(x))
```

```python
# we sample a sin(x) function
dx = np.pi/10
x = np.arange(0,2*np.pi,np.pi/10)

# we calculate the derivative, with np.gradient
plt.plot(x,np.gradient(np.sin(x), dx), '-*', label='approx')

# we compare it with the exact first derivative, i.e. cos(x)
plt.plot(x,np.cos(x), label='exact')
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def test(a, b, c, d=None):
    if c == 0:
        return a + b + c
    else:
        return a + b + c + np.sin(d)

a = 1
b = 2
c = 3

x = np.arange(1, 20, 1/100)
d = x
dx = 0.1
plt.plot(x,  np.gradient((lambda d: test(a, b, c, d))(x), dx))
```
